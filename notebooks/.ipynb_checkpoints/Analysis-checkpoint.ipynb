```python
# Performance Metrics:
# Completion Rate
total_test_clients = len(df_final[df_final['Variation'] == 'Test'])

confirm_test_clients = len(df_final[(df_final['Variation'] == 'Test') & (df_final['process_step'] == 'confirm')])

confirm_test_rate = confirm_test_clients / total_test_clients

total_control_clients = len(df_final[df_final['Variation'] == 'Control'])

confirm_control_clients = len(df_final[(df_final['Variation'] == 'Control') & (df_final['process_step'] == 'confirm')])

confirm_control_rate = confirm_control_clients / total_control_clients

print("Rate of clients from the test group that reached the step 'confirm': {:.2%}".format(confirm_test_rate))
print("Rate of clients from the control group that reached the step 'confirm': {:.2%}".format(confirm_control_rate))
```

```python
df_final['id'] = df_final.apply(lambda row: '_'.join(map(str, [row['client_id'], row['visitor_id'], row['visit_id']])), axis=1)
print(df_final)
```

```python
# Performance Metrics:
# Time Spent on Each Step
df_final['timestamp'] = pd.to_datetime(df_final['date_time'])

df_final.sort_values(by=['id', 'timestamp'], inplace=True)

def calculate_time_spent(group):
    group['time_spent'] = (group['timestamp'] - group['timestamp'].shift()).fillna(pd.Timedelta(seconds=0))
    return group

df_final = df_final.groupby('id').apply(calculate_time_spent)

df_filtered = df_final[(df_final['process_step'] != 'start')]

average_time_per_step = df_filtered.groupby(['Variation', 'process_step'])['time_spent'].mean()

print(average_time_per_step)
```

```python
# Performance Metrics:
# Error Rates
df_final.reset_index(drop=True, inplace=True)

df_final['is_error'] = df_final['process_step'].lt(df_final.groupby(df_final.index)['process_step'].shift().fillna(''))

test_group = df_final[df_final['Variation'] == 'Test']
test_error_rate = (test_group.groupby(test_group.index)['is_error'].sum() > 0).mean()

control_group = df_final[df_final['Variation'] == 'Control']
control_error_rate = (control_group.groupby(control_group.index)['is_error'].sum() > 0).mean()

print("Error Rate for the test group:", test_error_rate)
print("Error Rate for the control group:", control_error_rate)
```

```python
# Conclusion
# Based on the chosen KPIs, the new design does not performance that better compared to the old one. 
# The completion rate is only 2.26% higher, the time spent on each step is a bit higher, with the exception from the start until spet 1
```

```python
from statsmodels.stats.proportion import proportions_ztest

test_completion_rate = (test_group['process_step'] == 'confirm').mean()
control_completion_rate = (control_group['process_step'] == 'confirm').mean()

print("Completion Rate for the Test Group:", test_completion_rate)
print("Completion Rate for the Control Group:", control_completion_rate)

# Null hypothesis: There is no difference in completion rates between the test and control groups
# Alternative hypothesis: There is a difference in completion rates between the test and control groups

count = np.array([test_group['process_step'].eq('confirm').sum(), control_group['process_step'].eq('confirm').sum()])
nobs = np.array([len(test_group), len(control_group)])
z_stat, p_value = proportions_ztest(count, nobs)

alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis. There is a statistically significant difference in completion rates.")
else:
    print("Fail to reject the null hypothesis. There is no statistically significant difference in completion rates.")
```

```python
observed_completion_rate_increase = test_completion_rate - control_completion_rate

# Null hypothesis: Completion rate increase is less than 5%
# Alternative hypothesis: Completion rate increase is greater than or equal to 5%

count = np.array([test_group['process_step'].eq('confirm').sum(), control_group['process_step'].eq('confirm').sum()])
nobs = np.array([len(test_group), len(control_group)])
z_stat, p_value = proportions_ztest(count, nobs, alternative='larger')  

alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis. The completion rate increase meets or exceeds the 5% threshold.")
else:
    print("Fail to reject the null hypothesis. The completion rate increase does not meet the 5% threshold.")
```

```python
from scipy.stats import ttest_ind

new_process_age = df_final[df_final['Variation'] == 'Test']['clnt_age']
old_process_age = df_final[df_final['Variation'] == 'Control']['clnt_age']

t_stat, p_value = ttest_ind(new_process_age, old_process_age, equal_var=False)  

alpha = 0.05

if p_value < alpha:
    print("Reject the null hypothesis. The average age of clients engaging with the new process is different from those engaging with the old process.")
else:
    print("Fail to reject the null hypothesis. There is no significant difference in the average age of clients engaging with the new and old processes.")
```

```python
new_process_tenure = df_final[df_final['Variation'] == 'Test']['clnt_tenure_yr']
old_process_tenure = df_final[df_final['Variation'] == 'Control']['clnt_tenure_yr']

t_stat, p_value = ttest_ind(new_process_tenure, old_process_tenure, equal_var=False)

alpha = 0.05

if p_value < alpha:
    print("Reject the null hypothesis. The average client tenure of those engaging with the new process is different from those engaging with the old process.")
else:
    print("Fail to reject the null hypothesis. There is no significant difference in the average client tenure of those engaging with the new and old processes.")
```

```python
engagement_steps = ['confirm'] 

df_final['engaged'] = np.where(df_final['process_step'].isin(engagement_steps), 1, 0)

from scipy.stats import chi2_contingency

test_group_contingency = pd.crosstab(df_final[df_final['Variation'] == 'Test']['gendr'], df_final[df_final['Variation'] == 'Test']['engaged'])

control_group_contingency = pd.crosstab(df_final[df_final['Variation'] == 'Control']['gendr'], df_final[df_final['Variation'] == 'Control']['engaged'])

chi2_test_stat, p_value, _, _ = chi2_contingency(test_group_contingency)

alpha = 0.05

if p_value < alpha:
    print("Reject the null hypothesis. There is a difference in the proportions of engagement between genders in the test group.")
else:
    print("Fail to reject the null hypothesis. There is no significant difference in the proportions of engagement between genders in the test group.")

chi2_test_stat, p_value, _, _ = chi2_contingency(control_group_contingency)

if p_value < alpha:
    print("Reject the null hypothesis. There is a difference in the proportions of engagement between genders in the control group.")
else:
    print("Fail to reject the null hypothesis. There is no significant difference in the proportions of engagement between genders in the control group.")
```
